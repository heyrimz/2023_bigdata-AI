```python
import requests
# url 불러오기
response = requests.get(url+key+page+date+etc)
```

```python
#데이터 값 출력
contents = response.text
print(contents)
```

```python
import xml.etree.ElementTree as et
import bs4
from lxml import html

xml_obj = bs4.BeautifulSoup(contents,'lxml-xml')
rows = xml_obj.findAll('item')
print(rows)
```

```python
row_list = []
name_list = []
value_list = []
for i in range(len(rows)):
  cols = rows[i].find_all()
  for j in range(len(cols)):
    if i == 0:
      name_list.append(cols[j].name)
    value_list.append(cols[j].text)
  row_list.append(value_list)
  value_list = []

df = pd.DataFrame(row_list, columns = name_list)
print(df)
df.to_csv('환경부 국립환경과학원_미세먼지(금속성분) 실시간 정보'+ date + '.csv')
```
